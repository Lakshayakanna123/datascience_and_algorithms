```python
df = pd.read_csv('company.csv')
```

```python
inputs = df.iloc[:,:-1]
inputs
```

```python
target = df.iloc[:,-1:]
target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import *
from sklearn.metrics import r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs, target, test_size=0.8, random_state=1)
```

## Linear Reg

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
x_train_pred = reg.predict(x_train)
r2_score(y_train, x_train_pred)
```

```python
x_test_pred = reg.predict(x_test)
r2_score(y_test, x_test_pred)
```

## Lasso Reg

```python
las = Lasso()
```

```python
las.fit(x_train,y_train)
```

```python
x_test_pred_ls = las.predict(x_test)
r2_score(y_test, x_test_pred_ls)
```

## Ridge Reg

```python
rid = Ridge()
```

```python
rid.fit(x_train,y_train)
```

```python
x_test_pred_rd = rid.predict(x_test)
r2_score(y_test, x_test_pred_rd)
```
